```python
def automata(word):
    state = 0
    for letter in word:
        if state == 0:
            if letter.isdigit():
                state = 1
            elif letter == "_":
                state = 10
            elif letter.isalpha():
              state = 10
            else:
                raise Exception("-1")
        elif state == 1:
            if letter.isdigit():
                state = 1
            elif letter == '.':
                state = 2
            else:
                raise Exception("-1")
        elif state == 2:
            if letter.isdigit():
                state = 3
            else:
                raise Exception("-1")
        elif state == 3:
            if letter.isdigit():
                state = 3
            else:
                raise Exception("-1")
        elif state == 10:
            if letter.isalpha():
              state = 10
            elif letter == "_":
              state = 10
            elif letter.isdigit():
              state = 11
            else:
              raise Exception("-1")
        elif state == 11:
            if letter.isalpha():
              state = 11
            elif letter == "_":
              state = 11
            elif letter.isdigit():
              state = 11
            else:
              raise Exception("-1")
        else:
            raise Exception("-1")
    if state == 1:
        print("1")
    elif state == 3:
        print("2 ")
    elif state == 10:
        print("4")
    elif state == 11:
        print("4")
    else:
        raise Exception("-1")

automata("123456789")
automata("123.6")
try:
    automata("15a")
except Exception as e:
    print(e)
try:
    automata("")
except Exception as e:
    print(e)
try:
    automata("98.")
except Exception as e:
    print(e)
automata("_identificador__")
automata("___")
automata("D_a_n_i_e_l_21")
try:
    automata("Daniel.99")
except Exception as e:
    print(e)
```
